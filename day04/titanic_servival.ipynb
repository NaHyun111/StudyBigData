##### age 결측치(177건)를 치환

- 전체(891명)의 데이터에서 중앙값을 찾아서 결측치를 치환 

```python
titanic.describe()
```

```python
# 중앙값
median = titanic.age.median()
```

```python
titanic.age.mean()
```

```python
# 나이 결측치 치환
titanic['age'] = titanic['age'].fillna(median)
```

```python
titanic.isnull().sum()
```

```python
# 승선장소 확인
titanic.embarked.value_counts()
```

```python
# 승선장소 대부분이 S이기때문에 결측치 2개도 S로 할당
titanic.embarked = titanic.embarked.fillna('S')
```

```python
titanic.embark_town.value_counts()
```

```python
titanic.embark_town = titanic.embark_town.fillna('Southampton')
```

```python
## 승선 데크를 확인
titanic.deck.value_counts()
```

```python
titanic.deck = titanic.deck.fillna('C')
```

```python
# 결측치 여부 재조회
titanic.isnull().sum()
```

##### 성별에 따른 생존률 파이차트

```python
f, ax = plt.subplots(1, 2, figsize=(10,5))
titanic.survived[titanic.sex == 'male'].value_counts().plot.pie(autopct='%1.1f%%', ax=ax[0], explode=[0,0.05], shadow=True)

titanic.survived[titanic.sex == 'female'].value_counts().plot.pie(autopct='%1.1f%%', ax=ax[1], explode=[0,0.05], shadow=True)

ax[0].set_title('Surviced (Male)')
ax[1].set_title('Surviced (Female)')

plt.show()
```

##### 변수 사이 상관계수를 히트맵으로 시각화

```python
def category_age(x):  #연령대 가져오기 함수
    if x<10:
        return 0
    elif x<20:
        return 1
    elif x<30:
        return 2
    elif x<40:
        return 3
    elif x<50:
        return 4
    elif x<60:
        return 5
    elif x<70:
        return 6
    else:
        return 7
```

```python
titanic
```

```python
titanic['age_band'] = titanic['age'].apply(category_age)
```

```python
titanic['sex'] = titanic['sex'].map({'male':1, 'female':2})
```

```python
titanic['family'] = titanic['sibsp'] + titanic['parch'] + 1
```
